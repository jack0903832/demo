```python
x  =np.linspace(0,10,10)
plt.plot(x,np.sin(x),'-o')
plt.ylim(-1.5,1.5)
```

```python
x = np.linspace(0.05,10,100)
y = np.sin(x)
plt.plot(x,y,label="sin(x)")
plt.xlabel("x")
plt.ylabel("y")
#plt.rcParams['font.sans-serif'] = ['SimHei']
plt.title('')
plt.grid()
```

```python
x = np.linspace(0.05,10,100)
y = np.sin(x)
plt.ylim(-2,2)
plt.plot(x,y,'-o')
plt.axhline(y=0.8,ls='--',c='y')
```

```python
x = np.linspace(0.05,10,100)
y = np.sin(x)
plt.plot(x,y)
plt.axvspan(xmin=4,xmax=6,facecolor='r',alpha=0.3)
plt.axhspan(ymin=-0.2,ymax=0.2,facecolor='y',alpha=0.3)
```

```python
x = np.linspace(0.05,10,100)
y = np.sin(x)
plt.plot(x,y,'-o')
plt.text(3.2,0,"sin(x)",weight='bold',color='y')

```

```python
x = np.linspace(0.05,10,100)
y = np.sin(x)
plt.plot(x,y)
plt.annotate('MAX',xy=(np.pi/2, 1),xytext=(np.pi/2+1, 1),
             weight='bold',
             color='r',
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='r'))
```

```python
x = np.linspace(0,10,1000)
fig,ax = plt.subplots()
ax.plot(x,np.sin(x),label='sin')
ax.plot(x,np.cos(x),'--',label='cos')
ax.legend()
```
